```python
import os
import numpy as np
import tifffile as tiff
import matplotlib.pyplot as plt
import pandas as pd
from skimage.morphology import ball, dilation, erosion
from skimage.measure import label

def get_center_roi(shape, fraction=0.3):
    """
    Create a full-scale boolean mask with True in the central fraction of x and y.
    The mask has the same shape as the input image (assumed z, y, x).
    """
    z, y, x = shape
    start_y = int((1 - fraction) / 2 * y)
    end_y = start_y + int(fraction * y)
    start_x = int((1 - fraction) / 2 * x)
    end_x = start_x + int(fraction * x)
    roi = np.zeros((y, x), dtype=bool)
    roi[start_y:end_y, start_x:end_x] = True
    # Replicate roi for all z slices
    return np.repeat(roi[np.newaxis, :, :], z, axis=0)

def extract_central_core(full_core, center_roi):
    """
    From the full-scale core mask, keep only the connected components that overlap
    with the provided center_roi mask.
    """
    # Label the full core mask
    labeled = label(full_core, connectivity=1)
    # Find labels that occur in the central ROI
    labels_in_roi = np.unique(labeled[center_roi])
    # Remove background (label 0)
    labels_in_roi = labels_in_roi[labels_in_roi != 0]
    # Create new core mask containing only those labels
    new_core = np.isin(labeled, labels_in_roi)
    return new_core

def remove_non_touching_components(mask, reference_mask, selem):
    """
    Label connected components in mask and keep only those that touch the reference_mask.
    A dilation of the reference_mask is used to allow near-touching.
    """
    labeled = label(mask, connectivity=1)
    keep_mask = np.zeros_like(mask, dtype=bool)
    dilated_ref = dilation(reference_mask, selem)
    for region_label in np.unique(labeled):
        if region_label == 0:
            continue  # skip background
        component = (labeled == region_label)
        if np.any(component & dilated_ref):
            keep_mask |= component
    return keep_mask

def process_meshlike(full_meshlike, new_core, selem):
    """
    Process the full-scale meshlike mask by subtracting the new core mask and
    then removing connected components that do not touch the new core.
    (No extra dilation is applied to the meshlike mask itself.)
    """
    # Ensure meshlike is exclusive of core
    meshlike_exclusive = full_meshlike & (~new_core)
    new_meshlike = remove_non_touching_components(meshlike_exclusive, new_core, selem)
    return new_meshlike

def process_diffuse(full_diffuse, meshlike, selem):
    """
    Process the full-scale diffuse mask:
      - Dilate the diffuse mask,
      - Keep only those connected components that touch the meshlike mask,
      - Then erode to refine the boundaries.
    """
    dilated = dilation(full_diffuse, selem)
    new_diffuse = remove_non_touching_components(dilated, meshlike, selem)
    new_diffuse = erosion(new_diffuse, selem)
    return new_diffuse

def process_image_set(core_path, meshlike_path, diffuse_path, main_dir, base_name):
    # Load full-scale images (3D z-stack, shape: (z, y, x))
    core_img = tiff.imread(core_path)
    meshlike_img = tiff.imread(meshlike_path)
    diffuse_img = tiff.imread(diffuse_path)
    
    # Convert to boolean masks (nonzero pixels are True)
    full_core = core_img > 0
    full_meshlike = meshlike_img > 0
    full_diffuse = diffuse_img > 0
    
    shape = full_core.shape
    # Create central ROI mask (only in x-y)
    center_roi = get_center_roi(shape, fraction=0.3)
    
    # Extract central core: only keep connected core areas that overlap the central ROI
    new_core = extract_central_core(full_core, center_roi)
    
    # Define a structuring element (3D ball of radius 1)
    selem = ball(1)
    
    # Process meshlike at full scale: subtract core and remove regions not touching the core.
    new_meshlike = process_meshlike(full_meshlike, new_core, selem)
    
    # Process diffuse at full scale: dilate then filter by contact with meshlike, then erode.
    new_diffuse = process_diffuse(full_diffuse, new_meshlike, selem)
    
    # Plot a representative z-slice from the full-scale images.
    slice_idx = 14
    if slice_idx >= shape[0]:
        slice_idx = shape[0] // 2
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    axs[0].imshow(new_core[slice_idx], cmap='gray')
    axs[0].set_title(f'Core (Full) Slice {slice_idx}')
    axs[1].imshow(new_meshlike[slice_idx], cmap='gray')
    axs[1].set_title(f'Meshlike (Full) Slice {slice_idx}')
    axs[2].imshow(new_diffuse[slice_idx], cmap='gray')
    axs[2].set_title(f'Diffuse (Full) Slice {slice_idx}')
    plt.suptitle("Base File: " + base_name)
    plt.tight_layout()
    plt.show()
    
    # Save the new meshlike and diffuse full-scale masks as TIFF files.
    meshlike_save_path = os.path.join(main_dir, base_name + 'meshlike_new.ome.tiff')
    diffuse_save_path = os.path.join(main_dir, base_name + 'diffuse_new.ome.tiff')
    tiff.imwrite(meshlike_save_path, (new_meshlike.astype(np.uint8) * 255))
    tiff.imwrite(diffuse_save_path, (new_diffuse.astype(np.uint8) * 255))
    
    # Calculate volumes (number of positive voxels) in the full-scale processed masks.
    core_volume = np.sum(new_core)
    meshlike_volume = np.sum(new_meshlike)
    diffuse_volume = np.sum(new_diffuse)
    total_volume = core_volume + meshlike_volume + diffuse_volume
    
    return {
        'BaseName': base_name,
        'CoreVolume': core_volume,
        'MeshlikeVolume': meshlike_volume,
        'DiffuseVolume': diffuse_volume,
        'TotalVolume': total_volume
    }

def main():
    main_dir = '/Users/katherineridley/Projects/PlaqueStack/Subc/'
    files = os.listdir(main_dir)
    
    # Identify unique base file names by looking for files ending with "core.ome.tiff"
    base_names = []
    for file in files:
        if file.endswith("core.ome.tiff"):
            base = file.replace("core.ome.tiff", "")
            base_names.append(base)
    base_names = sorted(set(base_names))
    
    results = []
    
    for base in base_names:
        core_path = os.path.join(main_dir, base + "core.ome.tiff")
        meshlike_path = os.path.join(main_dir, base + "meshlike.ome.tiff")
        diffuse_path = os.path.join(main_dir, base + "diffuse.ome.tiff")
        
        # Check that all three mask files exist.
        if not (os.path.exists(core_path) and os.path.exists(meshlike_path) and os.path.exists(diffuse_path)):
            print(f"Missing one or more files for base {base}. Skipping.")
            continue
        
        print("Processing base file:", base)
        res = process_image_set(core_path, meshlike_path, diffuse_path, main_dir, base)
        results.append(res)
    
    # Save volume measurements to CSV (including the base file name).
    df = pd.DataFrame(results)
    csv_path = os.path.join(main_dir, "volumes_all.csv")
    df.to_csv(csv_path, index=False)
    print("Processing complete. Volume data saved to:", csv_path)

if __name__ == '__main__':
    main()
```

```python
import os
import pandas as pd

def main():
    # Set the main directory containing the CSV file.
    main_dir = '/Users/katherineridley/Projects/PlaqueStack/Subc/'
    csv_path = os.path.join(main_dir, "volumes_all.csv")
    
    # Read the CSV file into a DataFrame.
    df = pd.read_csv(csv_path)
    
    # List of volume columns to convert.
    volume_columns = ['CoreVolume', 'MeshlikeVolume', 'DiffuseVolume', 'TotalVolume']
    
    # Create new columns for per µm³ values by multiplying by 0.1.
    for col in volume_columns:
        new_col = col + "_um3"
        df[new_col] = df[col] * 0.1

    # Save the new DataFrame to a new CSV file.
    new_csv_path = os.path.join(main_dir, "volumes_all_um3.csv")
    df.to_csv(new_csv_path, index=False)
    print("New CSV saved to:", new_csv_path)

if __name__ == '__main__':
    main()
```
